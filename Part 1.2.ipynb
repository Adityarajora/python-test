```python
import nsepy as n
from datetime import *
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fun(s):
    if s=='INFY':
        infy = n.get_history(symbol='INFY', start=date(2015,1,1), end=date(2015,12,31))
        k=infy.copy()
        k['Rolling10']=infy['Close'].rolling(10).mean()
        k['Rolling75']=infy['Close'].rolling(75).mean()
        plt.plot(list(range(248)),k['Rolling10'],label="Rolling-10")
        plt.plot(list(range(248)),k['Rolling75'],label="Rolling-75")
        plt.legend()
        plt.xlabel('Stocks')
        plt.ylabel('Rolling window mean')
        
    else:
        tcs= n.get_history(symbol='TCS', start=date(2015,1,1), end=date(2015,12,31))
        k=tcs.copy()
        k['Rolling10']=tcs['Close'].rolling(10).mean()
        k['Rolling75']=tcs['Close'].rolling(75).mean()
        plt.plot(list(range(248)),k['Rolling10'],label="Rolling-10")
        plt.plot(list(range(248)),k['Rolling75'],label="Rolling-75")
        plt.legend()
        plt.xlabel('Stocks')
        plt.ylabel('Rolling window mean')
```

```python
fun('INFY')
```

```python
fun('TCS')
```
